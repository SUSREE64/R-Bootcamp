## R- Visualization with R-base


#### plot() Function with various options

#### plot(x,y, type, ----)
type
what type of plot should be drawn. Possible types are

>- "p" for points,

>- "l" for lines,

>- "b" for both,

>- "c" for the lines part alone of "b",

>- "o" for both ‘overplotted’,

>- "h" for ‘histogram’ like (or ‘high-density’) vertical lines,

>- "s" for stair steps,

>- "S" for other steps, see ‘Details’ below,

>- "n" for no plotting.
pch: the plotting symbol (default is open circle)

>- lty: the line type (default is solid line), can be dashed, dotted, etc.

>- lwd: the line width, specified as an integer multiple

>- col: the plotting color, specified as a number, string, or hex code; the colors() function gives you a vector of colors by name

>- xlab: character string for the x-axis label

>- ylab: character string for the y-axis label

#### point plot or scatter plot

### Plot annotation

```r
plot(x = a, y = b , type = 'b', pch = '$', col = 'green4') 
#pch = plotting character
#col = colour
# Adding a title with lables 
title(main = "Basic Point plot - main", sub = "sub Title", xlab = "x-lable", ylab = "Y-lable")
# Adding some text to the plot 
text(x = 2, y = 60, labels = "Added Blue text here", col = 'blue',cex = 1.5)
# cex is font size
#col is colour of the text
```

#### Histogram 

```r
# Histogram is a single variable plot that show the frequency of the variable
variable <- c(10,10,10,10, 20,20, 20,20, 30,30,30, 40,40,40, 50,50,50,50, 60,60,60,60,60,60,70,70,70,70,80,80,80,90,90,90,100,100,100)
hist(variable,breaks = 10,
     labels = TRUE,
     col = 'pink',
     border = 'blue',
     main = "Histogram Example",
     xlim = c(1,100),
     xlab = "Variable Values",
     ylab = 'Frequency of Var',
     ylim = c(0,10))
#col - Fill colour of the bars
#border - outline colour of the bars
#ylim = This should be a vector c(start, end) to define the y scale
#xlim = This should be a vector c(start, end) to define the x scale

```

### Pairs()
<p>If we have to plot multiple plots to see the co-relation of numerical values of the dataframe,
Pairs() plot is very handy to have a quick look at the variables and thier interdependency.</p>

```r
options(repr.plot.width=12, repr.plot.height=8)
pairs(formula = ~wt+mpg+disp+cyl,data = mtcars, main = "Scatterplot Matrix")
```

### barplot()


#### Basic Syantax of barplot
#### barplot(H,xlab,ylab,main, names.arg,col)
>- H is a vector or matrix containing numeric values used in bar chart.
>- xlab is the label for x axis.
>- ylab is the label for y axis.
>- main is the title of the bar chart.
>- names.arg is a vector of names appearing under each bar.
>-col is used to give colors to the bars in the graph.

```r
names <- c('sree','Raj','Ravi','Gita','Mahesh')
marks <- c(60, 78, 90, 74, 56)
options(repr.plot.width=12, repr.plot.height=5)
barplot(marks, 
        xlab = "Students Names",
        main = "Students Marks barplot",
       names.arg = names,
       xlim = c(0,6), 
       ylim = c(0,100), 
       ylab = "Marks",
       col = "#fad87a",
       border = "white")
# Adding the values of the barplot as text on the plot 
text(x = c(0.7,1.9,3.1,4.3,5.5),marks+5,labels=as.character(marks)) 
```

#### R - Pie charts
#### Syntax:
#### pie(x, labels, radius, main, col, clockwise)
>- x is a vector containing the numeric values used in the pie chart.
>- labels is used to give description to the slices.

>- radius indicates the radius of the circle of the pie chart.(value between −1 and +1).

>- main indicates the title of the chart.

>- col indicates the color palette.

>- clockwise is a logical value indicating if the slices are drawn clockwise or anti clockwise.

```r
names <- c('sree','Raj','Ravi','Gita','Mahi')
marks <- c(60, 78, 90, 74, 35)
percents <- round((marks/sum(marks))*100,1)
options(repr.plot.width=12, repr.plot.height=8)
pie(marks, radius = 1, main = "Marks Distrubtion", labels = percents, col = rainbow(length(marks)))
legend("topright", names, cex = 0.9, fill = rainbow(length(marks)))
   
```

### Mutiple plots as a grid side by side

```r
names <- c('sree','Raj','Ravi','Gita','Mahesh')
marks <- c(60, 78, 90, 74, 35)
percents <- round((marks/sum(marks))*100,1)
options(repr.plot.width=12, repr.plot.height=6)
#par(mflow = c(m,n)) m = Rows, n = Columns 
par(mfrow=c(1,2)) # set the plotting area into a 1*2 
barplot(height = marks, names.arg = names, main = "Barplot of Marks Distribution")
pie(marks, radius = 2, main = "PIE plot of Marks Distrubtion", labels = percents, col = rainbow(length(marks)))
```

```r
options(repr.plot.width=12, repr.plot.height=8)
par(mfrow=c(2,2)) # set the plotting area into a 1*2 
barplot(height = marks, names.arg = names, main = "Barplot of Marks Distribution")
pie(marks, radius = 1, main = "PIE plot of Marks Distrubtion", labels = percents, col = rainbow(length(marks)))
plot(y = marks, x = c(1,2,3,4,5), main = "Simple x-y plot", type = 'b', pch = 19, col="darkgreen", cex = 1.2)
pie(marks, radius = 1, main = "PIE plot of Marks Distrubtion", labels = percents, col = rainbow(length(marks)))
```

#### Multiple Plots Grid another example below

```r
## All plot options on type = ?. with a grid of plots 
par(mfrow=c(4,2))
options(repr.plot.width=12, repr.plot.height=8)
a <- 1:10
b <- a*a
plot(x = a, y = b , type = 'p', xlab = "with option p")
plot(x = a, y = b , type = 'l',xlab = "with option l")
plot(x = a, y = b , type = 'b', xlab = "with option b")
plot(x = a, y = b , type = 'c', xlab = "with option c")
plot(x = a, y = b , type = 'o', xlab = "with option o")
plot(x = a, y = b , type = 'h', xlab = "with option h")
plot(x = a, y = b , type = 's', xlab = "with option s")
plot(x = a, y = b , type = 'S', xlab = "with option S")
```
